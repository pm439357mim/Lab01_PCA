```python
import torch, torchvision
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# Przygotowanie danych, normalizacja MNIST
MNIST_MEAN, MNIST_STD = 0.1307, 0.3081
transform = torchvision.transforms.Compose([
    torchvision.transforms.ToTensor(),
    torchvision.transforms.Normalize((MNIST_MEAN,), (MNIST_STD,))
])

trainset = torchvision.datasets.MNIST("./data", train=True,
                                      download=True, transform=transform)
testset  = torchvision.datasets.MNIST("./data", train=False,
                                      download=True, transform=transform)

trainloader = DataLoader(trainset, batch_size=128, shuffle=True,  pin_memory=True)
testloader  = DataLoader(testset,  batch_size=1000, shuffle=False, pin_memory=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("device =", device)
```

```python
%%time
#  LeNet-5

class LeNet5(nn.Module):
    def __init__(self):
        super().__init__()

        self.conv1 = nn.Conv2d(in_channels=1,  out_channels=6,  kernel_size=5)   # 28×28  24×24
        self.conv2 = nn.Conv2d(in_channels=6,  out_channels=16, kernel_size=5)   # 12×12  8×8
        self.conv3 = nn.Conv2d(in_channels=16, out_channels=120, kernel_size=4)  #  4×4  1×1

        self.avgpool = nn.AvgPool2d(kernel_size=2, stride=2)

        self.fc1 = nn.Linear(120, 80)
        self.fc2 = nn.Linear(80,  10)


        self.dropout = nn.Dropout(0.05)
        self.relu    = nn.ReLU()

    def forward(self, x):

        x = self.relu(self.conv1(x))        # (B,6,24,24)
        x = self.avgpool(x)                 # (B,6,12,12)

        x = self.relu(self.conv2(x))        # (B,16,8,8)
        x = self.avgpool(x)                 # (B,16,4,4)

        x = self.relu(self.conv3(x))        # (B,120,1,1)
        x = x.squeeze(-1).squeeze(-1)       # (B,120)

        x = self.relu(self.fc1(x))          # (B,80)
        x = self.fc2(x)                     # (B,10)

        x = self.dropout(x)
        return x
```

```python
# trening
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Working on", device)

#parametry
epochs= 15
LR=  0.001
PY= 100

net= LeNet5().to(device)
optimizer = torch.optim.Adam(net.parameters(), lr=LR)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.3)
criterion  = nn.CrossEntropyLoss()

# pętla treningowa
net.train()
for epoch in range(1, epochs + 1):
    running = 0.0
    for batch, (x, y) in enumerate(trainloader, 1):
        x, y = x.to(device), y.to(device)

        optimizer.zero_grad()
        loss = criterion(net(x), y)
        loss.backward()
        optimizer.step()                     #aktualizacja wag

        running += loss.item()
        if batch % PY == 0:
            print("epoch:", epoch,
                  "batch:", batch,
                  "current batch loss:", running / PY)
            running = 0.0

    scheduler.step()


    net.eval()
    correct = 0
    with torch.no_grad():
        for x_test, y_test in testloader:
            x_test, y_test = x_test.to(device), y_test.to(device)
            preds = net(x_test).argmax(1)
            correct += (preds == y_test).sum().item()

    print("epoch:", epoch,
          "test accuracy:", correct / len(testset))
    net.train()                              # powrót przed kolejną epoką
```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML


MIN_VAL = -MNIST_MEAN / MNIST_STD
MAX_VAL = (1 - MNIST_MEAN) / MNIST_STD


steps1, lr1, frames1 = 1000, 0.01, 20
step1 = max(1, steps1 // frames1)

inp1 = torch.randn(10, 1, 28, 28, device=device, requires_grad=True)
opt1 = torch.optim.Adam([inp1], lr=lr1)
targets1 = torch.arange(10, device=device)

traj1, confs1 = [], []

for epoch in range(steps1):
    opt1.zero_grad()
    logits = net(inp1)
    loss   = F.cross_entropy(logits, targets1)
    loss.backward()
    opt1.step()
    inp1.data.clamp_(MIN_VAL, MAX_VAL)

    if epoch % step1 == 0:
        with torch.no_grad():
            traj1.append(inp1.detach().cpu().clone().numpy())
            probs = torch.softmax(logits, 1)
            confs1.append(
                probs[torch.arange(10), targets1].cpu().numpy()
            )


steps2, lr2, frames2 = 1000, 0.01, 20
step2 = max(1, steps2 // frames2)

inp2 = torch.randn(1, 1, 28, 28, device=device, requires_grad=True)
opt2 = torch.optim.Adam([inp2], lr=lr2)

traj2, confs2 = [], []

for epoch in range(steps2):
    opt2.zero_grad()
    logits2 = net(inp2)[0, [3, 8]]
    loss2   = -logits2.sum()
    loss2.backward()
    opt2.step()
    inp2.data.clamp_(MIN_VAL, MAX_VAL)

    if epoch % step2 == 0:
        with torch.no_grad():
            traj2.append(inp2.detach().cpu().clone().numpy())
            probs2 = torch.softmax(net(inp2), 1)
            confs2.append(probs2[0, [3, 8]].cpu().numpy())


lambdas = [0.001, 0.01, 0.1, 1.0, 10.0]
steps3, lr3, frames3 = 1000, 0.01, 20
step3 = max(1, steps3 // frames3)
targets3 = torch.arange(10, device=device)

all_traj3, all_confs3, all_dream3 = {}, {}, {}

for lam in lambdas:
    inp3 = torch.randn(10, 1, 28, 28, device=device, requires_grad=True)
    opt3 = torch.optim.Adam([inp3], lr=lr3)

    traj3, confs3 = [], []

    for epoch in range(steps3):
        opt3.zero_grad()
        logits3 = net(inp3)
        ce  = F.cross_entropy(logits3, targets3)
        l2  = lam * inp3.pow(2).mean()
        (ce + l2).backward()
        opt3.step()
        inp3.data.clamp_(MIN_VAL, MAX_VAL)

        if epoch % step3 == 0:
            with torch.no_grad():
                traj3.append(inp3.detach().cpu().clone().numpy())
                probs3 = torch.softmax(logits3, 1)
                confs3.append(
                    probs3[torch.arange(10), targets3].cpu().numpy()
                )

    all_traj3[lam]  = traj3
    all_confs3[lam] = confs3
    all_dream3[lam] = (
        (inp3 * MNIST_STD + MNIST_MEAN)
        .clamp(0, 1)
        .detach()
        .cpu()
        .numpy()
    )


    fig, ax = plt.subplots(2, 5, figsize=(10, 4))
    for a, im in zip(ax.flatten(), all_dream3[lam]):
        a.imshow(im[0], cmap='gray'); a.axis('off')
    plt.suptitle(f"Task 3 — L2  λ={lam}")
    plt.show()


def make_animation(traj, title, nrows, ncols, step):
    fig, axes = plt.subplots(nrows, ncols,
                             figsize=(ncols * 1.5, nrows * 1.5))
    axes = axes.flatten() if hasattr(axes, "flatten") else [axes]
    for ax in axes:
        ax.axis("off")
    ims = [
        ax.imshow(np.zeros((28, 28)), cmap="gray", vmin=0, vmax=1)
        for ax in axes
    ]

    def init():
        for im in ims:
            im.set_data(np.zeros((28, 28)))
        fig.suptitle(f"{title} – Epoch 0")
        return ims

    def animate(i):
        batch = (traj[i] * MNIST_STD + MNIST_MEAN).clip(0, 1)
        for im, frame in zip(ims, batch):
            im.set_data(frame[0])
        fig.suptitle(f"{title} – Epoch {(i + 1) * step}")
        return ims

    anim = FuncAnimation(
        fig, animate, frames=len(traj),
        init_func=init, blit=False, interval=300
    )
    plt.close(fig)
    return anim


anim1 = make_animation(traj1, "Task 1", 2, 5, step1)
anim2 = make_animation(traj2, "Task 2", 1, 1, step2)
anim3 = make_animation(all_traj3[0.1], "Task 3  λ=0.1", 2, 5, step3)

display(HTML(anim1.to_html5_video()))
display(HTML(anim2.to_html5_video()))
display(HTML(anim3.to_html5_video()))


def plot_confidence(confs, step, title, labels):
    arr = np.stack(confs)
    it  = np.arange(arr.shape[0]) * step
    plt.figure(figsize=(8, 4))
    for idx, lab in enumerate(labels):
        plt.plot(it, arr[:, idx], label=lab)
    plt.xlabel("Iteration")
    plt.ylabel("Confidence")
    plt.title(title)
    plt.legend(ncol=5)
    plt.show()

plot_confidence(confs1, step1,
                "Task 1 Confidence", [str(i) for i in range(10)])
plot_confidence(confs2, step2,
                "Task 2 Confidence", ["3", "8"])
plot_confidence(all_confs3[0.1], step3,
                "Task 3 Confidence  λ=0.1", [str(i) for i in range(10)])
```

```python
import torch
import random
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Working on", device)

#cyfry 0-9
samples_per_class = 9
picked = {k: [] for k in range(10)}

for img_batch, lbl_batch in testloader:
    for x, y in zip(img_batch, lbl_batch):
        if len(picked[y.item()]) < samples_per_class:
            picked[y.item()].append(x)
    if all(len(v) == samples_per_class for v in picked.values()):
        break

orig_imgs= torch.cat([torch.stack(picked[k]) for k in range(10)]).to(device)
orig_lbls= torch.cat([torch.full((samples_per_class,), k) for k in range(10)]).to(device)
B = orig_imgs.size(0)

# losujemy zły target
targets = torch.empty_like(orig_lbls)
for i, y in enumerate(orig_lbls):
    bad = list(range(10)); bad.remove(y.item())
    targets[i] = random.choice(bad)

#trained noise
lambdas = [10.0, 1.0, 0.1]
iters= 300
lr= 0.05

for p in net.parameters():
    p.requires_grad_(False)
net.eval()

results = []

for lam in lambdas:
    delta = torch.zeros_like(orig_imgs, requires_grad=True)
    opt= torch.optim.Adam([delta], lr=lr)

    for step in range(iters):
        opt.zero_grad()
        logits = net(orig_imgs + delta)
        loss = torch.nn.functional.cross_entropy(logits, targets) \
               + lam * delta.pow(2).mean()
        loss.backward()
        opt.step()
        with torch.no_grad():
            delta.clamp_(-1.0, 1.0)

    # skuteczność + pewność błędnej klasy
    with torch.no_grad():
        adv_logits = net(orig_imgs + delta)
        pred = adv_logits.argmax(1)
        success= (pred == targets).float().mean().item()
        delta_L2 = delta.view(B, -1).norm(dim=1).mean().item()
        wrong_conf = torch.softmax(adv_logits, dim=1)[torch.arange(B), targets].mean().item()

    results.append({"λ": lam,
                    "success": success,
                    "delta_norm": delta_L2,
                    "wrong_conf": wrong_conf})

#wizualizacja
    n_rows, n_cols = 10, samples_per_class
    fig, axes = plt.subplots(2 * n_rows, n_cols, figsize=(n_cols * 1.2, 2 * n_rows * 1.2))
    fig.suptitle(f"λ = {lam}   success = {success*100:.1f}%", fontsize=14)

    for cls in range(10):
        for j in range(samples_per_class):
            idx = cls * samples_per_class + j
# oryginał
            ax1= axes[2*cls, j]
            ax1.imshow(orig_imgs[idx].cpu().squeeze(), cmap='gray')
            ax1.axis('off')
            if j == 0:
                ax1.set_ylabel(f"{cls}", rotation=0, labelpad=12, va='center')

            ax2 = axes[2*cls + 1, j]
            ax2.imshow((orig_imgs + delta)[idx].detach().cpu().squeeze(), cmap='gray')
            ax2.axis('off')
            if j == 0:
                ax2.set_ylabel(f"→{pred[idx].item()}", rotation=0, labelpad=12, va='center')

    plt.tight_layout(rect=[0, 0, 1, 0.97])
    plt.show()


for r in results:
    print(f"lambda={r['λ']:>5} | success={r['success']*100:5.1f}% | "
          f"delta_norm={r['delta_norm']:.4f} | "
          f"avg wrong-class conf={r['wrong_conf']:.3f}")
```

Obrazy stworzone przez LeNet-5 są bardziej czytelne i przypominają prawdziwe cyfry z danych MNIST dla λ = 10 (są bardziej zbliżone do pisma ręcznego).
Na obrazkach z MLP jest ziarnisty szum, a w przypadku CNN (LeNet-5) bardziej rozmyty i gładki. Przy małych lambdach MLP cyfry są bardziej widoczne niż w przypadku CNN.

CNN ma inductive bias na lokalne krawędzie i przesuwane filtry 5x5.
Oraz gradient skleja takie lokalne fragmenty, przy maksymalizacji logitu danej klasy. Gdyż MLP nie tworzy takich struktur: dla kilku przypadkowych pikseli, aktywowuje odpowiednie neurony, tworząc ziarnisty szum.

Obrazy są gładkie i szum jest stłumiony przy λ=10. Po zmniejszeniu λ do 0,1, ziarniste artefakty zaczynają dominować.

Dla każdej z trzech wartości λ conf>0.9, sieć jest bardzo "pewna". Przy λ = 10 zakłócenie jest najmniejsze, a sieć myli się z pewnością około 0,92.
dla λ = 1 wokół pojawia się wyraźniejsza mgła, norma δ₂ rośnie, ale pewność błędnej klasy wzrasta do około 0,99.
Dla λ = 0,1 szum (ziarnistość na tle) jest dobrze widoczny, zakłócenie ma największą normę, ale cyfry da się rozpoznać, a sieć jest niemal pewna, że to zła klasa, czyli conf=0.998. Energia i widoczność szumu rosną. CNN łatwo zmienia decyzję nawet przy niewielkiej chaotycznej zmianie obrazu.
