```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
from torchvision import transforms
from torch.utils.data import DataLoader

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
# Przygotowanie danych, normalizacja MNIST
MNIST_MEAN, MNIST_STD = 0.1307, 0.3081
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((MNIST_MEAN,), (MNIST_STD,))
])
trainset = torchvision.datasets.MNIST(root="./data", train=True, download=True, transform=transform)
testset  = torchvision.datasets.MNIST(root="./data", train=False, download=True, transform=transform)
trainloader = DataLoader(trainset, batch_size=2048, shuffle=True)
testloader  = DataLoader(testset,  batch_size=1,    shuffle=False)
```

```python
# MLP (train)
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 1024), nn.ReLU(),
            nn.Linear(1024, 2048), nn.ReLU(),
            nn.Linear(2048, 256),  nn.ReLU(),
            nn.Linear(256, 10)
        )
        self.dropout = nn.Dropout(0.05)
    def forward(self, x):
        x = self.mlp(x)
        if self.training:
            x = self.dropout(x)
        return x
```

```python
# trening modelu
# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Working on {device}")

net = MLP().to(device)
optimizer = torch.optim.Adam(net.parameters(), 0.001)   #initial and fixed learning rate of 0.001.

net.train()    #it notifies the network layers (especially batchnorm or dropout layers, which we don't use in this example) that we are doing traning
for epoch in range(25):  #  an epoch is a training run through the whole data set

    for batch, data in enumerate(trainloader):
        batch_inputs, batch_labels = data

        batch_inputs = batch_inputs.to(device)  #explicitly moving the data to the target device
        batch_labels = batch_labels.to(device)

        #batch_inputs.squeeze(1)     #alternatively if not for a Flatten layer, squeeze() could be used to remove the second order of the tensor, the Channel, which is one-dimensional (this index can be equal to 0 only)

        optimizer.zero_grad()

        batch_outputs = net(batch_inputs)   #this line calls the forward(self, x) method of the MLP object. Please note, that the last layer of the MLP is linear
                                            #and MLP doesn't apply
                                            #the nonlinear activation after the last layer
        loss = torch.nn.functional.cross_entropy(batch_outputs, batch_labels, reduction = "mean") #instead, nonlinear softmax is applied internally in THIS loss function
        print("epoch:", epoch, "batch:", batch, "current batch loss:", loss.item())
        loss.backward()       #this computes gradients as we have seen in previous workshops
        optimizer.step()     #but this line in fact updates our neural network.
                                ####You can experiment - comment this line and check, that the loss DOE

# "zamro≈ºenie" wag dla "dreaming"
net.eval()
for p in net.parameters():
    p.requires_grad_(False)

```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.animation import HTMLWriter
from IPython.display import HTML



MIN_VAL = -MNIST_MEAN / MNIST_STD       # minimalna warto≈õƒá ((0 - mean)/std)
MAX_VAL = (1 - MNIST_MEAN) / MNIST_STD    # maksymalna warto≈õƒá ((1 - mean)/std)

#1:  0‚Äì9
steps1, lr1, frames1 = 1000, 0.01, 20
step1 = max(1, steps1 // frames1)
# szum Gaussa
inp1 = torch.randn(10,784, device=device, requires_grad=True)  # 10 sztuk
opt1  = torch.optim.Adam([inp1], lr=lr1)
targets1 = torch.arange(10, device=device) # etykiety 0‚Äì9

traj1, confs1 = [], [] # dla animacji

for epoch in range(steps1):
    opt1.zero_grad()
    logits = net(inp1.view(10,1,28,28)) # przez sieƒá
    loss   = F.cross_entropy(logits, targets1)  # cross‚Äêentropy loss
    loss.backward()   # liczymy gradien
    opt1.step()  # optymalizacja
    inp1.data.clamp_(MIN_VAL, MAX_VAL)  # warto≈õci wej≈õcia - zakres
     # zapis
    if epoch % step1 == 0:
        with torch.no_grad():
            traj1.append(inp1.view(10,1,28,28).cpu().numpy())
            probs = torch.softmax(logits,1)
            confs1.append(probs[torch.arange(10), targets1].cpu().numpy())

# 2 : 3 i 8
steps2, lr2, frames2 = 1000, 0.01, 20
step2 = max(1, steps2 // frames2)

inp2 = torch.randn(1,784, device=device, requires_grad=True)
opt2 = torch.optim.Adam([inp2], lr=lr2)

traj2, confs2 = [], []
for epoch in range(steps2):
    opt2.zero_grad()
    logits2 = net(inp2.view(1,1,28,28))[0,[3,8]]
    loss2   = -logits2.sum()  # maksymalizujemy sumƒô logit√≥w
    loss2.backward()
    opt2.step()
    inp2.data.clamp_(MIN_VAL, MAX_VAL)
    if epoch % step2 == 0:
        with torch.no_grad():
            traj2.append(inp2.view(1,1,28,28).cpu().numpy())
            probs2 = torch.softmax(net(inp2.view(1,1,28,28)),1)
            confs2.append(probs2[0,[3,8]].cpu().numpy())

# 3: L2
lambdas = [0.001, 0.01, 0.1, 1.0, 10.0]   # zakres
steps3, lr3, frames3 = 1000, 0.01, 20
step3 = max(1, steps3 // frames3)
targets3 = torch.arange(10, device=device)

all_traj3 = {}
all_confs3 = {}
all_dream3 = {}

for lam in lambdas:
    inp3 = torch.randn(10, 784, device=device, requires_grad=True)
    opt3 = torch.optim.Adam([inp3], lr=lr3)

    traj3, confs3 = [], []
    for epoch in range(steps3):
        opt3.zero_grad()
        logits3 = net(inp3.view(10,1,28,28))
        ce  = F.cross_entropy(logits3, targets3)
        l2  = lam * inp3.pow(2).mean()
        (ce + l2).backward()
        opt3.step()
        inp3.data.clamp_(MIN_VAL, MAX_VAL)

        if epoch % step3 == 0:
            with torch.no_grad():
                traj3.append(inp3.view(10,1,28,28).cpu().numpy())
                probs3 = torch.softmax(logits3, 1)
                confs3.append(probs3[torch.arange(10), targets3].cpu().numpy())

    #  wyniki
    all_traj3[lam]= traj3
    all_confs3[lam] = confs3
    all_dream3[lam] = (inp3.view(10,1,28,28)*MNIST_STD + MNIST_MEAN)\
                        .clamp(0,1).detach().cpu().numpy()

    # wizualizacja
    fig, ax = plt.subplots(2,5, figsize=(10,4))
    for a, im in zip(ax.flatten(), all_dream3[lam]):
        a.imshow(im[0], cmap='gray'); a.axis('off')
    plt.suptitle(f"Task 3 ‚Äî L2 Œª={lam}")
    plt.show()

#  animacja
def make_animation(traj, title, nrows, ncols, step):
    fig, axes = plt.subplots(nrows, ncols, figsize=(ncols*1.5, nrows*1.5))
    axes = axes.flatten() if hasattr(axes, 'flatten') else [axes]
    for ax in axes: ax.axis('off')
    ims = [ax.imshow(np.zeros((28,28)), cmap='gray', vmin=0, vmax=1) for ax in axes]
    def init():
        for im in ims: im.set_data(np.zeros((28,28)))
        fig.suptitle(f'{title} ‚Äì Epoch 0')
        return ims
    def animate(i):
        batch = (traj[i] * MNIST_STD + MNIST_MEAN).clip(0,1)
        for im, frame in zip(ims, batch):
            im.set_data(frame[0])
        fig.suptitle(f'{title} ‚Äì Epoch {(i+1)*step}')
        return ims
    anim = FuncAnimation(fig, animate, frames=len(traj),
                         init_func=init, blit=False, interval=300)
    plt.close(fig)
    return anim

anim1 = make_animation(traj1, 'Task 1', 2, 5, step1)
anim2 = make_animation(traj2, 'Task 2', 1, 1, step2)
anim3 = make_animation(traj3, 'Task 3 Œª=0.1', 2, 5, step3)

display(HTML(anim1.to_html5_video()))
display(HTML(anim2.to_html5_video()))
display(HTML(anim3.to_html5_video()))

# confidence i iteracje

def plot_confidence(confs, step, title, labels):
    arr = np.stack(confs)
    it  = np.arange(arr.shape[0]) * step
    plt.figure(figsize=(8,4))
    for idx, lab in enumerate(labels):
        plt.plot(it, arr[:,idx], label=lab)
    plt.xlabel("Iteration"); plt.ylabel("Confidence")
    plt.title(title)
    plt.legend(ncol=5)
    plt.show()

plot_confidence(confs1, step1, 'Task 1 Confidence', [str(i) for i in range(10)])
plot_confidence(confs2, step2, 'Task 2 Confidence', ['3','8'])
plot_confidence(confs3, step3, 'Task 3 Confidence Œª=0.1', [str(i) for i in range(10)])
```
